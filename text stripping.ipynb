```python
from bs4 import BeautifulSoup
from collections import Counter
import re
import requests
import pandas as pd
```

```python
url = 'https://www.periscopedata.com/data_analyst'
response = requests.get(url)
html_doc= response.text
soup = BeautifulSoup(html_doc)
print soup.prettify()
```

```python
listings= soup.findAll(name='div', attrs={'class':'person'})
P= re.findall(r'\b\w+', str(listings))
Q=Counter(P)
sorted(Q.items(), key=lambda x: x[1], reverse=True)
```

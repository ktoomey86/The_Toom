# Modeling # 

```python
import os
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import StandardScaler
```

```python
x_standard = StandardScaler().fit_transform(predictors)
cov_mat = np.cov(x_standard.T)
eigenValues, eigenVectors = np.linalg.eig(cov_mat)
# print(eigenValues)
# print "\n"
# print(eigenVectors)
```

```python
eig_pairs = [(np.abs(eigenValues[i]), eigenVectors[:,i])\
             for i in range(len(eigenValues))]

eig_pairs
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
pca.fit(x_standard)

# print("The information (explained variance) \
# contained in each principal component: ", pca.explained_variance_ratio_)

# print "\n"

# print(pca.components_)

```

```python
tot = sum(eigenValues)
var_exp = [(i / tot)*100 for i in sorted(eigenValues)]
cum_var_exp = np.cumsum(var_exp)

print(cum_var_exp)
```
